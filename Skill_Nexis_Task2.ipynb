```python
embark=pd.get_dummies(house_data['parking'],drop_first=True )
embark.head(5)
```

```python
house_data=pd.concat([house_data,embark],axis=1)
house_data.head(5)
```

```python
#Train and Test Data
x=house_data.drop('parking',axis=1)
y=house_data['parking']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler

x_train_unscaled, x_test_unscaled, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)

# Apply StandardScaler to the features
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train_unscaled)
x_test = scaler.transform(x_test_unscaled)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=1000) # Increased max_iter for convergence
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
import numpy as np
import matplotlib as plt
import pandas as pd
%matplotlib inline
dataset=pd.read_csv('Titanic-Dataset.csv')
dataset.head(2)
```

```python
x=dataset.iloc[:,[2,4]].values
y=dataset.iloc[:,4].values
x
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
import numpy as np

sc = StandardScaler()
le = LabelEncoder()
x_train_processed = np.column_stack((x_train[:, 0].astype(float), le.fit_transform(x_train[:, 1])))
x_test_processed = np.column_stack((x_test[:, 0].astype(float), le.transform(x_test[:, 1])))

x_train = sc.fit_transform(x_train_processed)
x_test = sc.transform(x_test_processed)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)

y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
accuracy_score(y_test,y_pred)*100
```
